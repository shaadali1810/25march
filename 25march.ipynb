```python
df1=df1
```

```python
df1['Installs']=df1['Installs'].str.replace('+',' ')
```

```python
df1['Installs']=df1['Installs'].str.replace(',','0')
df1['Installs']=df1['Installs'].str.replace('Free','0')
df1['Installs']=df1['Installs'].astype(float)
```

```python
df_app=df1.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
```

```python
df_app.Installs=df_app.Installs/100000
```

```python
df2=df_app.head(10)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Installs",y='Category',data=df2)
```
